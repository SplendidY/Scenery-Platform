```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import random
import requests
import os
```

```python
sceneries_df = pd.read_excel("zhejiang_sceneries.xlsx")
names = sceneries_df['name']
names
```

```python
driver = webdriver.Chrome()

for name in names[518:]:
    #输入搜索内容
    url = "https://image.baidu.com/search/index?tn=baiduimage&ps=1&ct=201326592&lm=-1&cl=2&nc=1&ie=utf-8&dyTabStr=MCwzLDEsMiw2LDQsNSw4LDcsOQ%3D%3D&word=" + name
    driver.get(url)
    driver.implicitly_wait(10)

    # search = driver.find_element(By.ID,'search')
    # searchform = search.find_element(By.XPATH,'//div[@class="s_search"]/div[@id="imgSearchBox"]/div[@id="imgMainSearch"]/div[@class="s_nav"]/form[@class="searchform"]')
    # word_input = searchform.find_element(By.XPATH,'//span[@class="s_ipt_wr"]/input[@id="kw"]')
    # word_button = searchform.find_element(By.XPATH,'//span[@class="s_btn_wr"]')
    # word_input.send_keys(Keys.CONTROL + 'a')
    # word_input.send_keys(Keys.DELETE)
    # word_input.send_keys(name)
    # word_button.click()
    # driver.implicitly_wait(10)
    time.sleep(random.uniform(3,5))

    #读取图片url并下载
    imgid = driver.find_element(By.XPATH,'/html/body/div[@id="wrapper"]/div[@class="cg_avatar"]/div[@id="imgid"]')
    imgpage= imgid.find_element(By.CLASS_NAME,'imgpage')

    
    imglist = imgpage.find_element(By.TAG_NAME,'ul')
    images = imglist.find_elements(By.TAG_NAME,'li')
    for image in images:
        url_of_picture = image.get_attribute('data-objurl')
        try:
            response = requests.get(url_of_picture)
            response.raise_for_status()
        except requests.exceptions.HTTPError as err:
            continue
        except Exception as err:
            continue
        with open("pictures/" + name + ".jpg", 'wb') as file:
            file.write(response.content)
        print("成功下载图片")
        break
```
